### Univariate for Numerical Features

```python
num_cols = ['reting','ram','internal_memory','battery_size',
            'charging_speed','rear_mp','front_mp',
            'display_size','refresh_rate','proccessor_speed']
for col in num_cols:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

### Univariate for Categorical Features

```python
cat_cols = ['brand','processor_brand','os',
            'is_5g','is_nfc','is_ir_blaster','fast_charge']
for col in cat_cols:
    sns.countplot(x=df[col])
    plt.xticks(rotation=45)
    plt.title(col)
    plt.show()
```
