###  Handle Missing Values in `is_nfc` Column

In this step, missing values in the **`is_nfc`** column are filled using the **mode** (most frequent value) of the columrue)


```python
df1['is_nfc'].value_counts()
```

```python
df1['is_nfc'].fillna(0,inplace=True)
```

```python
df1.isnull().sum()
```

### Handle Missing Values in reting Column (Brand-wise Mean)

In this step, missing values in the reting column are filled using the mean rating of the corresponding brand.

```python
df1['reting'].isnull().sum()
```

```python
df1[df1['reting'].isnull()]
```

```python
round(df1.groupby('brand')['reting'].mean()['Samsung'],2)
```

```python
df1['reting'] = df1.groupby('brand')['reting'].transform(lambda x:x.fillna(round(x.mean(),2)))
```

```python
df1.isnull().sum()
```

### Handle Missing Values in ram Column (Mode Imputation)

In this step, missing values in the ram column are filled using the mode (most frequent value) of the column.

```python
most_frequent_value=df['ram'].value_counts().sort_values(ascending=False).index[0]
```

```python
df1['ram'].fillna(most_frequent_value).isnull().sum()
```

```python
df1['ram'].fillna(most_frequent_value,inplace=True)
```

```python
df1.isnull().sum()
```

### Handle Missing Values in os Column (Brand-based Imputation)

In this step, missing values in the os column are filled based on the corresponding brand:

Apple → iOS

All other brands → Android

```python
df1['os'].value_counts()
```

```python
df1['brand'].value_counts()
```

```python
df1[df1['os'].isna()]
```

```python
df1[df['brand']=='Apple']['os'].fillna('a')
```

```python
def fill_os(row):
    if pd.isna(row['os']):
        if row['brand'] == 'Apple':
            return 'iOS'
        else:
            return 'Android'
    return row['os']

df1['os']=df1.apply(fill_os, axis=1)
```

```python
df1.isnull().sum()
```

```python
df1.to_csv("cleaned_dataset.csv",index=False)
```
